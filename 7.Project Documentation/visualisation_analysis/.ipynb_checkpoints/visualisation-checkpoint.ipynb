```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("credit_card.csv")
df.head()
```

```python
df = pd.read_csv("../Dataset/application_record.csv")
```

```python
df.info()
df.describe()
```

```python
df.isnull().sum()
```

```python
plt.figure(figsize=(8,5))
plt.hist(df["AMT_INCOME_TOTAL"], bins=30)
plt.title("Income Distribution")
plt.xlabel("Annual Income")
plt.ylabel("Number of Applicants")
plt.savefig("income_distribution.png")
plt.show()
```

```python
plt.figure(figsize=(8,5))
df["NAME_EDUCATION_TYPE"].value_counts().plot(kind="bar")
plt.title("Education Type Distribution")
plt.xlabel("Education")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("education_distribution.png")
plt.show()
```

```python
plt.figure(figsize=(8,5))
df["NAME_HOUSING_TYPE"].value_counts().plot(kind="bar")
plt.title("Housing Type Distribution")
plt.xlabel("Housing Type")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("housing_distribution.png")
plt.show()
```

```python
numeric_df = df.select_dtypes(include=["int64", "float64"])

plt.figure(figsize=(10,6))
sns.heatmap(numeric_df.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.tight_layout()
plt.savefig("correlation_heatmap.png")
plt.show()
```
